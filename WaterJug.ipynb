```python
import random
import string
```

```python
class node:
    graph = {}
    def __init__ (self,name,valueGiven,parentID):
        self.nameOfNode = name
        self.value = valueGiven
        self.parentOfNode = parentID
        self.ChildrenList = []
        node.graph[name] = {}
        node.graph[name]["Nodename"] = name
        node.graph[name]["Value"] = valueGiven
        node.graph[name]["ParentID"] = self.parentOfNode.nameOfNode
        node.graph[name]["Childrens"] = []
        
    def fillA(self):
        if self.value[0] != 3:
            Childname = randomNameGenerator()
            node.graph[self.nameOfNode]["Childrens"].append(Childname)
            Childname = node(Childname,[3,self.value[1]],self)
            self.ChildrenList.append(Childname)
            
    def fillB(self):
        if self.value[1] != 4:
            Childname = randomNameGenerator()
            node.graph[self.nameOfNode]["Childrens"].append(Childname)
            Childname = node(Childname,[self.value[0],4],self)
            self.ChildrenList.append(Childname)
            
    def emptyA(self):
        if self.value[0] != 0:
            Childname = randomNameGenerator()
            node.graph[self.nameOfNode]["Childrens"].append(Childname)
            Childname = node(Childname,[0,self.value[1]],self)
            self.ChildrenList.append(Childname)
            
    def emptyB(self):
        if self.value[1] != 0:
            Childname = randomNameGenerator()
            node.graph[self.nameOfNode]["Childrens"].append(Childname)
            Childname = node(Childname,[self.value[0],0],self)
            self.ChildrenList.append(Childname)
    
    def transfer_B_to_A(self):
        if self.value[1] != 0:
            Childname = randomNameGenerator()
            node.graph[self.nameOfNode]["Childrens"].append(Childname)
            Childname = node(Childname,[self.value[0],self.value[1]],self)
            while Childname.value[0] < 3 and Childname.value[1] > 0:
                Childname.value[1] = Childname.value[1] - 1 
                Childname.value[0] = Childname.value[0] + 1
            self.ChildrenList.append(Childname)
            
    def transfer_A_to_B(self):
        if self.value[0] != 0:
            Childname = randomNameGenerator()
            node.graph[self.nameOfNode]["Childrens"].append(Childname)
            Childname = node(Childname,[self.value[0],self.value[1]],self)
            while Childname.value[0] > 0 and Childname.value[1] < 4:
                Childname.value[0] = Childname.value[0] - 1 
                Childname.value[1] = Childname.value[1] + 1
            self.ChildrenList.append(Childname)
            
    def childrenShow(self):
        listToShow = [s.value for s in self.ChildrenList]
        return listToShow
```

```python
# to do 6 listed operations on given node and add resulting nodes into its child list

def doOperations(currentNode):
    currentNode.fillA()
    currentNode.fillB()
    currentNode.emptyA()
    currentNode.emptyB()
    currentNode.transfer_B_to_A()
    currentNode.transfer_A_to_B()
    return currentNode.ChildrenList

# to generate random name for nodes


def randomNameGenerator():
    res = ''.join(random.choices(string.ascii_uppercase + string.digits, k = 5)) 
    return res

#built tree and check if it is visited
def tree(listForOp):
    for i in listForOp:
        if i.value not in visited:
            visited.append(i.value)
            if i.value == [0,2]:
                print("Got It")
                print(backtrack(i))
            else:
                print("Node in consideration",i.value)
                listForOp2 = doOperations(i)
                print("Children : ", i.childrenShow())
                print("Visited : ", visited)
                tree(listForOp2) #recursive function
                
def backtrack(currentNode):
    path = []
    while currentNode.parentOfNode.nameOfNode != "root1":
        path = currentNode.value
        currentNode = currentNode.parentOfNode
    return path
            
```
